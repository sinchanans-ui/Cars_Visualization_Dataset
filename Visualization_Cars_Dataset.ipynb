# Tasks To Be Performed:

```python
fig=plt.subplots(2,2,figsize=(20,10))
plt.subplot(2,2,1)
sns.scatterplot(x=data['mpg'],y=data['hp'],data=data)   #sns.scatterplot
plt.xlabel("mpg")
plt.ylabel("hp")
plt.title("mpg-hp")

plt.subplot(2,2,2)
plt.scatter(x=data['mpg'],y=data['weightlbs'])    #plt.scatter
plt.xlabel("mpg")
plt.ylabel("weightlbs")
plt.title("mpg-weight")

plt.subplot(2,2,3)
sns.scatterplot(x=data['hp'],y=data['mpg'],data=data)
plt.xlabel("hp")
plt.ylabel("mpg")
plt.title("hp-mpg")

plt.subplot(2,2,4)
sns.scatterplot(x=data['hp'],y=data['time-to-60'],data=data)
plt.xlabel("hp")
plt.ylabel("time-to-60")
plt.title("hp-time-to-60")
```

```python
#y-->hp,mpg,weight,time-to-60
#x-->cylinders
#hue---->brand
```

```python
#a
fig=plt.subplots(2,2,figsize=(20,10))
plt.subplot(2,2,1)
sns.barplot(x=data['cylinders'],y=data['hp'],data=data,hue=data['brand'])

plt.subplot(2,2,2)
sns.barplot(x=data['cylinders'],y=data['mpg'],data=data,hue=data['brand'])

plt.subplot(2,2,3)
sns.barplot(x=data['cylinders'],y=data['weightlbs'],data=data,hue=data['brand'])

plt.subplot(2,2,4)
sns.barplot(x=data['cylinders'],y=data['time-to-60'],data=data,hue=data['brand'])
```

```python
#b
#a
fig=plt.subplots(2,2,figsize=(20,10))
plt.subplot(2,2,1)
sns.barplot(x=data['year'],y=data['hp'],data=data,hue=data['brand'])

plt.subplot(2,2,2)
sns.barplot(x=data['year'],y=data['mpg'],data=data,hue=data['brand'])

plt.subplot(2,2,3)
sns.barplot(x=data['year'],y=data['weightlbs'],data=data,hue=data['brand'])

plt.subplot(2,2,4)
sns.barplot(x=data['year'],y=data['time-to-60'],data=data,hue=data['brand'])
```
